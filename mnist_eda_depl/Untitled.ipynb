```python
import boto3
s3 = boto3.client("s3", region_name='ap-south-1')
s3.download_file("kubeflow-ds-test-raw-lake", "data/train.csv", "data/train.csv")
```

```python
import pandas as pd
df_train = df = pd.read_csv('data/train.csv', sep=',', header=0)

x = df_train.iloc[:, 1:]
y = df_train['label'].tolist()

# # Select 10000 rows data as a testing dataset
x_test = x.iloc[0:10000, :].values.astype('float32') # all pixel values 
y_test = y[0:10000] # Select label for testing data
x_train = x.iloc[10000:, :].values.astype('float32') # all pixel values 
y_train = y[10000:]

pd.DataFrame(x_train).to_csv('training_data.csv', index=False)
pd.DataFrame(y_train).to_csv('training_labels.csv', index=False)

pd.DataFrame(x_test).to_csv('test_data.csv', index=False)
pd.DataFrame(y_test).to_csv('test_labels.csv', index=False)
print("ALL DONE")
```

```python
import pandas as pd
    
x_train = pd.read_csv('training_data.csv')
x_test = pd.read_csv('test_data.csv')

x_train = x_train/255.0
x_test = x_test/255.0

x_train.to_csv('normalized_train_data.csv', index=False)
x_test.to_csv('normalized_test_data.csv', index=False)

print("ALL DONE")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

depth = 1
n_estimators = 50
random_state = 5


x_train = pd.read_csv('normalized_train_data.csv')
y_train = pd.read_csv('training_labels.csv', index_col=False)['0'].tolist()


x_test = pd.read_csv('normalized_test_data.csv')
y_test = pd.read_csv('test_labels.csv', index_col=False)['0'].tolist()


model_clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=depth, random_state=random_state)

# Train the Random Forest algorithm
model_clf.fit(x_train, y_train)


y_pred = model_clf.predict(x_test)

# calculate accuracy
total = len(y_pred)
wrong = 0
for pred, truth in zip(y_pred, y_test):

    wrong += 1 if pred != truth else 0
print("wrong: ", wrong)
accuracy = ((total - wrong)/total) * 100.0

print("Accuracy:", accuracy)
########## log model ####################
import mlflow

mlflow.set_tracking_uri(mlflow_tracking_uri)
mlflow.set_experiment(mlflow_experiment_name)

with mlflow.start_run() as run:
    mlflow.log_param("n_estimators", n_estimators)
    mlflow.log_param("max_depth", depth)
    mlflow.log_param("random_state", random_state)
    mlflow.log_metric("accuracy", accuracy)

    mlflow.sklearn.log_model(model_clf, "model", registered_model_name=model_name)

    model_artifact_location = run.info.artifact_uri + "/model"
    logged_model = 'runs:/'+run.info.run_id+'/model'

print(accuracy, logged_model)
```
